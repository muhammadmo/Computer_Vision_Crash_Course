```python
assets = pd.read_csv("AMZN.csv")
assets
```

```python
assets = assets.sort_index(ascending=True)
assets
```

```python
#select column
assets = assets[["Date", "Close"]]
assets
```

```python
#reset index
assets.reset_index(drop=True)
```

```python
#delete missing values
assets = assets.dropna()
assets
```

```python
#rolling
assets["new_col"] = assets[["Close"]].rolling(3).mean()
assets
```

```python
#shift
assets["Close"].shift(1)
```

```python
#slicing
assets.iloc[0:500,0:10]
```

```python
#slicing by condition
assets.loc[0:18]
#assets.loc[0:18,["Date"]]




#condition by value
#assets.loc[assets["Close"]>600]
```
