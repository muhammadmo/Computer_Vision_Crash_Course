```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generate random data
arr = np.random.randint(0,15,size=(15,))

square = arr**2

mut = arr*3

print(arr)
print(square)
print(mut)
```

```python
#plot data
plt.plot(arr)
plt.plot(square)
plt.plot(mut)
```

```python
plt.plot(arr)
plt.plot(square)
plt.plot(mut)

#change x&y label
plt.xlabel("Random Number")
plt.ylabel("values")

#change the title
plt.title("Graph")

#Put a Legend
plt.legend(["Normal", "Square", "Multplication"])

#show the chart
plt.show()
```

```python
#point clouds
plt.scatter(arr, mut)
plt.scatter(arr, square)

plt.xlabel("Random Number")
plt.ylabel("values")

plt.title("Scatter")

plt.legend(["Square", "Exp"])

plt.show()
```

```python
#adapt the size
plt.figure(figsize=(15,8))

#plot the line
plt.plot(arr, alpha=0.3)
#plt.plot(arr, alpha=0.8, color="red")

#plot the line
plt.plot(square, "-o", linewidth=3)

#plot the line
plt.plot(mut, "o")


plt.ylabel("values")

plt.title("Graph")

plt.legend(["Normal", "Square", "Multplication"])

plt.show()
```
